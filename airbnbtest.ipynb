```python
cache_dataimport streamlit as st
import pandas as pd
import pydeck as pdk

# Load the Airbnb dataset
@st.cache
def load_data():
    data = pd.read_csv('Airbnb.csv')
    data = data[['latitude', 'longitude', 'price', 'review_scores', 'room_type', 'name', 'number_of_reviews']]
    data = data.dropna(subset=['latitude', 'longitude', 'price'])
    return data

# Load data
st.title("Airbnb Geospatial Visualization")
st.sidebar.title("Filter Options")
data = load_data()

# Sidebar Filters
price_range = st.sidebar.slider("Price Range", int(data['price'].min()), int(data['price'].max()), (50, 200))
review_score = st.sidebar.slider("Minimum Review Score", 0, 100, 50)
room_type = st.sidebar.multiselect("Room Type", options=data['room_type'].unique(), default=data['room_type'].unique())

# Filter data
filtered_data = data[
    (data['price'] >= price_range[0]) & (data['price'] <= price_range[1]) &
    (data['review_scores'] >= review_score) &
    (data['room_type'].isin(room_type))
]

# Map Visualization
st.subheader("Interactive Map of Airbnb Listings")
st.write(f"Showing {len(filtered_data)} listings")
map_view = pdk.Deck(
    map_style="mapbox://styles/mapbox/light-v9",
    initial_view_state=pdk.ViewState(
        latitude=filtered_data['latitude'].mean(),
        longitude=filtered_data['longitude'].mean(),
        zoom=10,
        pitch=50
    ),
    layers=[
        pdk.Layer(
            "ScatterplotLayer",
            data=filtered_data,
            get_position=["longitude", "latitude"],
            get_color="[200, 30, 0, 160]",
            get_radius=200,
            pickable=True,
        ),
    ],
    tooltip={"text": "{name}\nPrice: ${price}\nReview Score: {review_scores}"}
)
st.pydeck_chart(map_view)

# Tabular Data
st.subheader("Filtered Listings")
st.dataframe(filtered_data[['name', 'price', 'review_scores', 'room_type', 'number_of_reviews']])
```
